```python

#เรียกใช้งานฟังค์ชั่น
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
#lap01
sin_sig = thinkdsp.SinSignal(freq=7013, amp=13.0, offset=0)
sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#lap02
signal =  thinkdsp.TriangleSignal(4000)
duration = signal.period*2500
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
kittikonwave=signal.make_wave(duration,framerate=10000)
kittikonwave.apodize()
kittikonwave.make_audio()
```

```python
spectrum=kittikonwave.make_spectrum()
spectrum.plot()
```

```python
signal =  thinkdsp.SquareSignal(4000)
duration = signal.period*2500
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05,1.05],legend=False)
segment.plot()
```

```python
kdwave = signal.make_wave(duration, framerate=10000)
kdwave.apodize()
kdwave.make_audio()
```

```python
spectrum=kdwave.make_spectrum()
spectrum.plot()
```
